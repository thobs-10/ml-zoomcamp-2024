```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input, GlobalAveragePooling2D
from tensorflow.keras.applications.xception import preprocess_input
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.applications import Xception
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
def create_cnn_model(input_shape=(200, 200, 3)):
    input_layer = Input(shape=input_shape)
    base_model = Xception(
        weights='imagenet',
        include_top=False,
        input_tensor=input_layer
    )
    base_model.trainable = False
    x = GlobalAveragePooling2D()(base_model.output)
    x = Dense(64, activation='relu')(x)
    output = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=input_layer, outputs=output)
    optimizer = SGD(learning_rate=0.002, momentum=0.8)
    model.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
model = create_cnn_model()
```

```python
model.summary()
```

```python
import os

data_dir = 'data'

def get_generators(data_dir):
  train_datagen = ImageDataGenerator(rescale=1./255)
  test_datagen = ImageDataGenerator(rescale=1./255)

  train_generator = train_datagen.flow_from_directory(
      os.path.join(data_dir, 'train'),
      target_size=(200, 200),
      batch_size=20,
      class_mode='binary',
      shuffle=True
  )

  test_generator = test_datagen.flow_from_directory(
      os.path.join(data_dir, 'test'),
      target_size=(200, 200),
      batch_size=20,
      class_mode='binary',
      shuffle=True
  )
  return train_generator, test_generator
```

```python
train_generator, test_generator = get_generators(data_dir)
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=test_generator
)
```

```python
def augmented_model(input_shape=(200, 200, 3)):
  train_datagen = ImageDataGenerator(
          rescale=1./255,
          rotation_range=50,
          width_shift_range=0.1,
          height_shift_range=0.1,
          zoom_range=0.1,
          horizontal_flip=True,
          fill_mode='nearest'
  )

  test_datagen = ImageDataGenerator(rescale=1./255)

  train_generator = train_datagen.flow_from_directory(
      os.path.join(data_dir, 'train'),
      target_size=(200, 200),
      batch_size=20,
      class_mode='binary',
      shuffle=True
  )

  test_generator = test_datagen.flow_from_directory(
      os.path.join(data_dir, 'test'),
      target_size=(200, 200),
      batch_size=20,
      class_mode='binary',
      shuffle=True
  )
  return train_generator, test_generator

train_generator, test_generator = augmented_model(data_dir)
history = model.fit(
    train_generator,
    epochs=20,
    validation_data=test_generator
)
```
